```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import os
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
merge = df1.merge(df2, how = 'inner', on = 'Date', left_index = True)
```

```python
merge.head(5)
```

```python
headlines = []
for row in range(0, len(merge.index)):
  headlines.append(' '.join( str(x) for x in merge.iloc[row, 2:27]))
```

```python
headlines[0]
```

```python
# Clean the data
clean_headlines = []
for i in range(0, len(headlines)):
  clean_headlines.append(re.sub("b[(')]", ' ', headlines[i]))
  clean_headlines[i] = re.sub('b(")', ' ', clean_headlines[i])
  clean_headlines[i] = re.sub("\'", ' ', clean_headlines[i])
```

```python
# show the combined clean headlines
clean_headlines[20]
```

```python
# adding the clean headlines to the merged dataset
merge['Combined_News'] = clean_headlines
merge['Combined_News'][0]
```

```python
# show the first 5 rows of the merged datasets
merge.head(5)
```

```python
# Create the function for subjectivity and polarity
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity 
def getPolarity(text):
  return TextBlob(text).sentiment.polarity  
```

```python
# Creating the columns for subjectivity and polarity
merge['Subjectivity'] = merge['Combined_News'].apply(getSubjectivity)
merge['Polarity'] = merge['Combined_News'].apply(getPolarity)
```

```python
# show the new columns in the merge datasets
merge.head(5)
```

```python
# Create a function to get the sentiment scores
def getSIA(text):
  sia = SentimentIntensityAnalyzer()
  sentiment = sia.polarity_scores(text)
  return sentiment
```

```python
# Get the sentiment scores for each day
compound = []
neg = []
pos = []
neu = []
SIA = 0

for i in range(0, len(merge['Combined_News'])):
  SIA = getSIA(merge['Combined_News'][i])
  compound.append(SIA['compound'])
  neg.append(SIA['neg'])
  neu.append(SIA['neu'])
  pos.append(SIA['pos'])
```

```python
# Store the sentiment scores in the merge data set
merge['Compound'] = compound
merge['Negative'] = neg
merge['Neutral'] = neu
merge['Positive'] = pos
```

```python
merge.head(5)
```

```python
keep_columns = ['Open', 'High', 'Low', 'Volume', 'Subjectivity', 'Compound', 'Negative', 'Neutral', 'Positive', 'Label']
df = merge[keep_columns]
df
```

```python
# Create the feature dataset
x = df
x = np.array(x.drop(['Label'], 1))
# Create the target dataset
y = np.array(df['Label'])
```

```python
# According to the famous financial data rules 
# Create a data set with 80% traning data and 20% test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#Create and train the model
model = LinearDiscriminantAnalysis().fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
y_test
```

```python
#Showing the model metrics
print( classification_report(y_test, predictions))
```
